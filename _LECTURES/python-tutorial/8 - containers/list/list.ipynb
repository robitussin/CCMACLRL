# List

Lists are used to store multiple items in a single variable.

List items are ordered, changeable, and allow duplicate values.

```python
# Declaring a list
numbers = [29, 58, 66, 71, 87]
```

Using the **list()** constructor

```python
# note the double round-brackets
thislist = list(("apple", "banana", "cherry"))
print(thislist)
```

## List Length

To determine how many items a list has, use the len() function:

```python
len(numbers)
```

## Access list items

```python
# Accessing an element
numbers[2]
```

```python
mylist = ['December 23, 2015', 'Bread Gloves', 8.51]

date = mylist[0]
item = mylist[1]
price = mylist[2]

print(date)
```

```python
date, item, price = ['December 23, 2015', 'Bread Gloves', 8.51]

print(date)
```

```python
# Negative Indexing
# -1 refers to the last item
print(numbers[-1])
```

```python
# Check if item exists
if 66 in numbers:
  print("Yes, 66 is in the numbers list")
```

## Change List items

```python
# Updating an element
numbers[2] = 68
numbers
```

```python
# Change a range of Items
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
thislist[1:3] = ["blackcurrant", "watermelon"]
print(thislist)
```

```python
# Change two items with one value
thislist = ["apple", "banana", "cherry"]
thislist[1:3] = ["watermelon"]
print(thislist)
```

## Add List items

### ***.append()*** to add item at the end of the list

```python
# Adding an element
thislist = ["apple", "banana", "cherry"]
thislist.append("orange")
print(thislist)
```

### ***.insert()*** to add item at a specific index

```python
thislist = ["apple", "banana", "cherry"]
thislist.insert(1, "orange")
print(thislist)
```

## Range of Indexes/Slicing

You can specify a range of indexes by specifying where to start and where to end the range.

When specifying a range, the return value will be a new list with the specified items.

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
thislist[:2]
```

```python
thislist[:2] = [0,0]
```

```python
thislist
```

## Remove List Items

```python
thislist[:2] = []
```

```python
thislist
```

#### Clear entire list

```python
thislist[:] = []
```
